```python

# load dataset
# specify the delimiter as '\t' (tab) as the data might be tab-separated.
df = pd.read_csv('/content/bbc-news-data.csv', delimiter='\t', on_bad_lines='skip')

# print the column names to verify.
print(df.columns)
```

```python

# access the columns in dataset based on the output
data = df['category'] + ' ' + df['filename'] + ' ' +  df['title'] + ' ' + df['content']
data = data.tolist()
```

```python

# initialize TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.8, min_df=0.1)

# fit and transform the corpus
tfidf_matrix = vectorizer.fit_transform(data)

# get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# create a df to store TF-IDF values
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# find the most important words (with highest TF-IDF values) for each article
for i in range(len(data)):
  top_words = df_tfidf.iloc[i].sort_values(ascending=False).head(10)
  print(f"Article {i+1}:")
  print(top_words)
  print("\n")
```

```python
# group articles by category since there were too many individual aticles
grouped = df.groupby('category')
```

```python
# initialize TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.8, min_df=0.1)
```

```python
# iterate over each category
for name, group in grouped:
  # merge text data for the current category
  data = group['category'] + ' ' + group['filename'] + ' ' +  group['title'] + ' ' + group['content']
  data = data.tolist()

  # fit and transform the corpus for the current category
  tfidf_matrix = vectorizer.fit_transform(data)

  # get feature names (words)
  feature_names = vectorizer.get_feature_names_out()

  # create a df to store TF-IDF values for the current category
  df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

  # calculate average TF-IDF values for each word in the category
  average_tfidf = df_tfidf.mean()

  # find the top 10 words with highest average TF-IDF values for the category
  top_words = average_tfidf.sort_values(ascending=False).head(10)

  # print the top words for the current category
  print(f"Category: {name}")
  print(top_words)
  print("\n")
```
